```python
df = pd.read_csv('employees.csv')
df['Senior Management'] = df['Senior Management'].astype("bool")
df['Gender'] = df['Gender'].astype("category")
df.head()
```

##### Compare the accuracy of a condition with "==" : returns a boolean series
- Filer a dataframe by creating a mask

```python
df[df['Gender'] == 'Male']
```

##### Not Equal Conditional Statement
- use the filter to return everything except the identified value

```python
mask = df['Team'] != 'Marketing'
df[mask].head(5)
```

##### Conditional Statement Between 2 Values: 

- must include the UPPER & LOWER condition in parentheses 
- the qualifier is placed in the middle

```python
mask = (df['Salary'] > 1000) & (df['Salary'] < 40000)
df[mask].head(7)
```

##### Conditional Statements with Dates
- Dates must be a datetime object for conditional statements to operate correctly

```python
df['Start Date'] = pd.to_datetime(df['Start Date'])
df['Start Date']
```

```python
mask = df['Start Date'] < '8/6/1993'
df[mask]
```

##### Filter a period between 2 dates

```python
mask = (df['Start Date'] > '1/1/1980') & (df['Start Date'] < '1/1/1985')
df[mask]
```

##### Multiple Conditions use multiple Masks

```python
mask = df['Gender'] == 'Male'
mask1 = df['Team'] == 'Marketing'
df[mask & mask1].head(3)
```

```python
mask1 = df['Senior Management'] == False
mask2 = df['Start Date'] < "1990-01-01"
df[mask1 | mask2]
```

##### isin() method finds multiple values returns boolean 
- Doesnt work with datetimes but with strings

```python
df['Start Date'].isin(['2010-07-22', '2012-03-08', '981-10-23'])
```
