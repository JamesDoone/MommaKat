##### Delete a row or column from a dataframe
##### default = 0 or rows
##### MUST SORT THE INDEX IN ORDER TO DROP

```python
bond = pd.read_csv('jamesbond.csv', index_col='Film')
bond.head(5)
```

```python
bond.sort_index()
```

```python
bond.drop(["Dr. No", "Casino Royale", "Live and Let Die"])
```

##### Drop Columns
##### Must change axis=0 to axis=1 or axis='columns'

```python
bond.drop('Director', axis='columns')
```
