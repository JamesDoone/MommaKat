##### Iterating thru Groups

```python
fort = pd.read_csv('fortune1000.csv', index_col='Rank')
sector = fort.groupby('Sector')
fort.head(3)
```

##### Create a dataframe to store the extracted data

```python
df = pd.DataFrame(columns= fort.columns)
df
```

##### step 2: loop thru groupby to extract data and append it to the df we created
- whenever you iterating over a groupby object, must provide 2 temporary variable names.
- 1st variable = must represent the group
- 2nd variable = must represent the actual dataframe
    - the dataframe for each sector will be stored in the temporary variable called data

```python
for sector, data in sectors:
    print(sector, data)
    highest_rev_inGroup = data.nlargest(1, "Revenue")
    df = df.append(highest_rev_inGroup) #must reassign df because there is no inplace method
```

```python
df
```

##### 2nd Example

```python
cities = fort.groupby("Location")
df = pd.DataFrame(columns = fort.columns)
df
```

```python
for city, data in cities:
    highest_rev_inCity = data.nlargest(1, "Revenue")
    df = df.append(highest_rev_inCity)
```
