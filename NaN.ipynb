##### dropna() will drop any ROWS with any NaN values.
##### if there is 1 NaN the ENTIRE  row will be dropped
##### The parameter 'how=any' is the default. 
##### 'how='all'" will ONLY drop a row if the ENTIRE row is populated with NaN values

```python
df.dropna()
```

```python
df.dropna(how='all')
```

##### To make changes permanent, inplace = True

##### To remove a column that has a NaN, change axis = 1

```python
df.dropna(axis=1)
```

##### subset() parameter allows the rows with NaN values to be dropped for targeted columns

```python
df.dropna(subset=['Salary'])
```

##### Fill in Null Values with the .fillna() Method
##### Does not work with categorical variables
##### To avoid this problem--call on the column as a series and perform the operation with inplace=True
##### Must identify value that will replace NaN values

```python
df['Salary'].fillna(0, inplace=True)
```

```python
df
```

##### Categorical variables: fillna() by calling on the column and by providing 'the replacement value'

```python
df['College'].fillna("HS_Recruit", inplace=True)
```

```python
df
```

##### The .isnull() and notnull() Methods
##### Returns boolean values

```python
%ls
```

```python
df = pd.read_csv('employees.csv', parse_dates=["Start Date", "Last Login Time"])
df["Senior Management"] = df['Senior Management'].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.head(3)
```

```python
df.info()
```

##### isnull() identifies NaN values by returning a boolean series

```python
df['Team'].isnull()
```

##### Create a mask to pull NaN rows

```python
mask = df['Team'].isnull()
```

```python
df[mask]
```

##### notnull()

```python
df['Team'].notnull()
```

##### notnull() in a mask will return a dataframe with no NaNs

```python
mask = df['Team'].notnull()
df[mask]
```
