##### Sort a Series
##### Convert a dataframe into a series with usecols=['col_name'] and squeeze=True

##### When numbers are strings with a symbol 
##### replace() the symbol
##### convert string to float with astype() to enable sort_values()

```python
df = pd.read_csv('chicago.csv', usecols=["Employee Annual Salary"], squeeze=True)
df.head()
```

##### replace() '$' with ''

```python
df = df.str.replace('$', '')
df.head()
```

```python
df = df.astype(float)
df
```

##### sort_values()
##### default: ascending=True or lowest numerals ascending to highest numerals
##### letters: 'A' ascending to 'Z'

```python
df.sort_values()
```

##### ascending = False will sort laregst numerals first descending to lowest numerals
##### letters: Z to A

```python
df.sort_values(ascending=False)
```

##### Sort a DataFrame with the .sort_values() Method

```python
%ls
```

```python
nba = pd.read_csv('nba.csv')
nba.head()
```

##### the 'by=' parameter identifies which column will be sorted

```python
height = nba.sort_values(by='Height')
height.head(10)
```

```python
nba.sort_values(by='College')
```

##### NaN values when the sort_values parameter is passed
- na_position = 'last'

```python
salary = nba.sort_values('Salary')
salary.head()
```

##### sort_values() by 2 objects instead of one
##### to control how each colummn is sorted adjust the ascending parameter for each column

```python
salary_height = nba.sort_values(by=['Salary', 'Height'], ascending=[False, False])
salary_height.head(10)
```
