##### Truncate

```python
%matplotlib inline
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
tickers = ['TSLA', 'GOOG']
start = "2017-01-01"
myData = pd.DataFrame()
for ticker in tickers:
    myData[ticker] = wb.DataReader(ticker, data_source='google',start=start)['Close']
```

```python
myData
```

```python
# (mydata/mydata.iloc[0]*100).plot(figsize=(16,12))
(myData/myData.iloc[0]*100).plot(figsize=(16,12))
```

##### Truncate()

```python
myData.truncate(before="2017-05-01", after="2017-06-13")
```
