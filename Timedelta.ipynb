##### Timedelta is an object of change in time

```python
import pandas as pd
```

```python
time_B = pd.Timestamp("2013-02-01")
time_C = pd.Timestamp("2013-02-13")
```

```python
time_C - time_B
```

##### Add minutes and seconds and it will be included in the delta

```python
time_B = pd.Timestamp("2013-02-01 12:09 PM")
time_C = pd.Timestamp("2013-02-13 1:24 AM")
```

```python
time_C - time_B
```

##### Create a timedelta object

```python
pd.Timedelta(days= 3, minutes=29, hours=11)
```
